The columns are masked to hide personal information. The label column in this set named 'Class', which shows whether or not a row represents a fraudulent transaction. The values can either be a 1 or a 0, and so we have a classification problem on our hands.

I will change the “Class” variable to a factor. I learned that this improves performance when using this variable for machine learning as it will be treated as a truly categorical variable. It works because we know the elements can only be one of two discrete values, 0 for non-fraud, 1 for fraud.

I also decided to remove the 'Time' column because it is just time since the first transaction- not the absolute time of each transaction.

```python
creditcard_dataset = creditcard_dataset.drop(['Time'], axis=1)
creditcard_dataset['Class'] = pd.factorize(creditcard_dataset['Class'])[0]

```

```python
count_classes = creditcard_dataset.value_counts(creditcard_dataset['Class'], sort = True)
count_classes.plot(kind = 'bar', rot=0)
plt.title("Label distribution")
plt.xlabel("Fraud Label")
plt.ylabel("Frequency");
```

```python
creditcard_dataset['Class'].value_counts()
```

This data is quite skewed as far as the labeled class goes, and so we would likely have a big bias problem in any model we make due to the sheer number of records in the "not fraud" class.

To remedy this we can use over-sampling, which is a way of synthetically increasing the number of records in the under-represented class. I use the SMOT library to do this.

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Separating features and target
X = creditcard_dataset.drop('Class', axis=1)
y = creditcard_dataset['Class']

# Splitting data into train and test using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Initializing SMOTE and balancing the training data- its important to only balance the training data set and not the testing data set
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Visualize the balanced data
plt.figure(figsize=(10,5))
y_train_smote.value_counts().plot(kind='bar', title='Fraud class histogram after SMOTE')
plt.xlabel('Fraud')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
```

```python
# Create the random forest classifier and train it on the newly separated training data
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
# Create a set of predictions by passing the test data through the newly created model.
y_pred = rf.predict(X_test)

# A quick look at the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Tree Visualisation
# We can actually look a small portion of some of the decision trees used to create the random forest.
from sklearn.tree import export_graphviz
from IPython.display import Image
import graphviz

for i in range(3):
    tree = rf.estimators_[i]
    dot_data = export_graphviz(tree,
                               feature_names=X_train.columns,
                               filled=True,
                               max_depth=2,
                               impurity=False,
                               proportion=True)
    graph = graphviz.Source(dot_data)
    display(graph)
```

```python
# A quick look at the other performance metrics

y_pred = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

I chose to use the random forest algorithm for the credit card fraud dataset for two reasons: The classification problem lends itself to using a decision tree approach, while the size of the data made it necessary to use a random forest of trees to reduce training time (even then it took a few minutes). I also chose this method so I can somewhat see what decision boundaries the algorithm used by displaying the individual decision trees- I thought that was cool. Only a few columns in this data set are strongly correlated in any direction, and so on a re-run I would optimize this by removing the columns that dont seem to be affecting the outcome very much. I would then experiment with the max_depth and n_estimators variables that are set to certain defaults by scikitlearn. n_estimators refers to how many individual decision trees are created, and the more trees the better performance a model will have, at the expense of higher training time. After removing the unneeded columns I might be able to afford the extra trees. The more interesting variable to me is the max_depth, which by default is None- "the nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples." (From the scikit learn docs here: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html). Having "too many" layers can lead to overfitting, while too little can lead to underfitting. I don't suspect either of those since the model performed well with the test data, but it is good to know that the parameter can be tweaked if need be. I think this model could still perform well given less data because the trees are going to be formed by randomly selecting data from the original data set. We can get around the problem of less data by increasing the num_estimators.
